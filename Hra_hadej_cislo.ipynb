```python
import random
import os

SUBOR_SKORE = "skore.txt"

def nacitaj_skore():
    skore = []
    if os.path.exists(SUBOR_SKORE):
        with open(SUBOR_SKORE, "r") as f:
            for riadok in f:
                meno, pokusy = riadok.strip().split(",")
                skore.append((meno, int(pokusy)))
    return skore

def uloz_skore(meno, pokusy):
    with open(SUBOR_SKORE, "a") as f:
        f.write(f"{meno},{pokusy}\n")

def zobraz_tab_skor():
    skore = nacitaj_skore()
    if not skore:
        print("游닔 Tabu쬶a sk칩re je zatia pr치zdna.")
        return

    print("\n游끤 TOP hr치캜i:")
    zoradene = sorted(skore, key=lambda x: x[1])[:5]
    for poradie, (meno, pokusy) in enumerate(zoradene, 1):
        print(f"{poradie}. {meno} - {pokusy} pokusov")

def hadaj_dvaja_hraci():
    print("游꿢 Vitaj v hre H치da캜ka 캜칤sla pre 2 hr치캜ov!")
    meno1 = input("Zadaj meno hr치캜a 1: ")
    meno2 = input("Zadaj meno hr치캜a 2: ")

    tajne_cislo = random.randint(1, 100)
    pokusy1 = 0
    pokusy2 = 0
    hrac_na_rade = 1

    while True:
        try:
            if hrac_na_rade == 1:
                tip = int(input(f"{meno1}, h치daj 캜칤slo medzi 1 a 100: "))
                pokusy1 += 1
                if tip == tajne_cislo:
                    print(f"\n游꿀 {meno1} vyhr치va! Uh치dol 캜칤slo {tajne_cislo} za {pokusy1} pokusov.")
                    uloz_skore(meno1, pokusy1)
                    break
                elif tip < tajne_cislo:
                    print("游댷 Sk칰s vy코코ie 캜칤slo.")
                else:
                    print("游댶 Sk칰s ni쮄멸e 캜칤slo.")
                hrac_na_rade = 2
            else:
                tip = int(input(f"{meno2}, h치daj 캜칤slo medzi 1 a 100: "))
                pokusy2 += 1
                if tip == tajne_cislo:
                    print(f"\n游꿀 {meno2} vyhr치va! Uh치dol 캜칤slo {tajne_cislo} za {pokusy2} pokusov.")
                    uloz_skore(meno2, pokusy2)
                    break
                elif tip < tajne_cislo:
                    print("游댷 Sk칰s vy코코ie 캜칤slo.")
                else:
                    print("游댶 Sk칰s ni쮄멸e 캜칤slo.")
                hrac_na_rade = 1
        except ValueError:
            print("丘멆잺 Zadaj platn칠 cel칠 캜칤slo.")

    zobraz_tab_skor()

if __name__ == "__main__":
    hadaj_dvaja_hraci()
```
