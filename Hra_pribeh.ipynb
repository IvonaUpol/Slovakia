```python
# prompt: Napiš mi příběh a dej mi na konci 2 možnosti na výběr. Po výběru vytvoř další část příběhu. Vždy je jen jedna možnost správná.

# Tvorba hry
# Příběh s dvěmi možnostmi na konci. Po výběru se vytvoří další část příběhu. Vždy je jen jedna možnost správná.

import ipywidgets as widgets
from IPython.display import display

def display_story_part(text, options, correct_option_index, next_story_part_func=None):
    """Displays a part of the story with interactive options."""
    print(text)

    if not options:
        print("\nKonec příběhu.")
        return

    buttons = []
    for i, option in enumerate(options):
        button = widgets.Button(description=option)
        button.on_click(lambda b, index=i: handle_choice(index, correct_option_index, next_story_part_func))
        buttons.append(button)

    display(widgets.VBox(buttons))

def handle_choice(chosen_index, correct_index, next_story_part_func):
    """Handles the user's choice and proceeds if correct."""
    # Clear previous output to keep the notebook clean
    from IPython.display import clear_output
    clear_output()

    if chosen_index == correct_index:
        print("Správná volba!\n")
        if next_story_part_func:
            next_story_part_func()
    else:
        print("Nesprávná volba. Zkus to znovu.")
        # You might want to redisplay the current part or end the game here
        # For simplicity, let's just indicate it was wrong and stop this path.
        # To replay, you'd need to re-run the cell.
        print("Konec hry.")


# --- Příběh ---

def part1():
    story_text = """
Probudil ses v temné místnosti. Nic si nepamatuješ. Jediná věc, kterou cítíš, je studená podlaha pod sebou.
V dálce slyšíš slabé kapání vody. Vstaneš a snažíš se zorientovat.
Před tebou jsou dvoje dveře. Jedny vlevo, jedny vpravo.
Které si vybereš?
    """
    options = ["Dveře vlevo", "Dveře vpravo"]
    correct_option_index = 0 # Vlevo je správně
    display_story_part(story_text, options, correct_option_index, part2)

def part2():
    story_text = """
Vybral sis dveře vlevo. Otevřely se s nepříjemným skřípáním.
Za nimi se nachází dlouhá, úzká chodba osvětlená slabým, blikajícím světlem.
Na konci chodby vidíš stín. Připomíná postavu. Co uděláš?
    """
    options = ["Pomalu se přiblížíš ke stínu", "Zkusíš se vrátit do předchozí místnosti"]
    correct_option_index = 0 # Přiblížit se je správně
    display_story_part(story_text, options, correct_option_index, part3)

def part3():
    story_text = """
Rozhodl ses pomalu přiblížit. Jakmile se dostaneš blíž, zjistíš, že stín patří starému muži sedícímu na zemi.
Vypadá vyčerpaně, ale jeho oči jsou bystré. Ukáže na malou krabičku vedle sebe.
"Otevři ji," zašeptá chraplavým hlasem.
Máš dvě možnosti.
    """
    options = ["Otevřeš krabičku", "Zeptej se ho, co je uvnitř"]
    correct_option_index = 0 # Otevřít krabičku je správně
    display_story_part(story_text, options, correct_option_index, part4)

def part4():
    story_text = """
S obavami otevřeš krabičku. Uvnitř najdeš malý, lesklý klíč.
Starý muž se usměje. "Ten klíč otevře dveře k východu. Ještě nejsi v bezpečí, ale teď máš šanci."
Ukáže na další dveře na konci chodby. Vypadají masivně.
Vzal jsi klíč a ...
    """
    # End of this path for now. You can add more parts.
    display_story_part(story_text, [], -1, None) # Empty options means end of this path

# --- Spustit hru ---
part1()
```
